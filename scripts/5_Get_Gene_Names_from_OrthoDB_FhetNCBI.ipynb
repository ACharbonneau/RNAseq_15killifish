```python
missing = []
for i in geneID:
    if i.startswith("ENS"):
        a = ODB8_OG.loc[ODB8_OG['protein_id'] == i]
        ind = len(a.index.values)
        b = 0
        while b < ind :
            og8 = a.at[a.index.values[b], 'odb8_og_id']
            b+=1
            c = ODB9_OG.loc[ODB9_OG['ODB8'] == og8]
            if c.empty:
                print("Not found:",og8)
                missing.append(og8)
            else:
                og9 = c.at[c.index.values[0],'ODB9']
                command_parts = ['wget ','http://www.orthodb.org','/group?','id=',og9,' -O ',i,'.orthodb']
                command = ''.join(command_parts)
                s = subprocess.Popen(command, shell=True)
                s.wait()
                eog_filename = i+'.orthodb'
                with open(eog_filename) as r:
                    lines = r.readlines()
                    j = json.loads(lines[0])
                    data = j["data"]
                    name = data["name"]
                    print(name)
```
