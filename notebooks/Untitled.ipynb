```python
import os
import pandas as pd
import numpy as np
```

```python
counts_table_file = "/Users/johnsolk/Documents/UCDavis/Whitehead/kfish_salmon/Ensembl_15killifish_normcounts_nostats.csv"
```

```python
counts = pd.read_csv(counts_table_file,index_col=0)
```

```python
counts.head()
```

```python
counts = counts + 1
log2counts = np.log2(counts)
```

```python
colnames = list(log2counts.columns.values)
species_treatment_cols = {}
species_list = []
for i in colnames:
    species = i.split("_")[0] + "_" + i.split("_")[1]
    if species not in species_list:
        species_list.append(species)
    species_treatment = i.split("_")[0] + "_" + i.split("_")[1] + "_" + i.split("_")[2]
    if species_treatment in species_treatment_cols:
        species_treatment_cols[species_treatment].append(i)
    else:
        species_treatment_cols[species_treatment] = [i]
print(species_treatment_cols)
```

```python
for species_treatment in species_treatment_cols:
    cols = species_treatment_cols[species_treatment]
    newcolname = "Mean_"+species_treatment
    log2counts[newcolname] = log2counts[cols].mean(axis=1)
```

```python
log2counts.head()
```

```python
# Contrast1 = Mean_species_transfer - Mean_species_FW
# Contrast2 = Mean_species_BW - Mean_species_FW
# Contrast3 = Mean_species_transfer - Mean_species_BW
colnames = list(log2counts.columns.values)
contrast1df = pd.DataFrame()
contrast2df = pd.DataFrame()
contrast3df = pd.DataFrame()

for species in species_list:
    print(species)
    transfer = "Mean_" + species + "_transfer"
    FW = "Mean_" + species + "_FW"
    BW = "Mean_" + species + "_BW"
    contrast1 = transfer + "-" + FW
    contrast2 = BW + "-" + FW
    contrast3 = transfer + "-" + BW
    if transfer in colnames and FW in colnames:
        log2counts[contrast1] = log2counts[transfer] - log2counts[FW]
        contrast1df[contrast1] = log2counts[transfer] - log2counts[FW]
    else:
        print("Missing contrast1",contrast1)
    if BW in colnames and FW in colnames:
        log2counts[contrast2] = log2counts[BW] - log2counts[FW]
        contrast2df[contrast2] = log2counts[BW] - log2counts[FW]
    else:
        print("Missing contrast2",contrast2)
    if transfer in colnames and BW in colnames:
        log2counts[contrast3] = log2counts[transfer] - log2counts[BW]
        contrast3df[contrast3] = log2counts[transfer] - log2counts[BW]
    else:
        print("Missing contrast3",contrast3)
```
