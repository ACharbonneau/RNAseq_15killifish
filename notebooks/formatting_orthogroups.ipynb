# change colnames in 'Orthogroups.csv' to remove .pep?

```python
species = list(orthogroups.columns.values)
```

```python
print(species)
```

```python
species_new = ["" if x=="Unnamed: 0" else x for x in species]
```

```python
print(species_new)
```

```python
species_new_new = [x.split(".")[0] if x.endswith(".pep") else x for x in species_new]
```

```python
print(species_new_new)
```

```python
species_new_new = [x[:1] + '_' + x[1:] if x!="" else x for x in species_new_new]
```

```python
print(species_new_new)
```

```python
orthogroups.columns = species_new_new
```

```python
new_orthogroups_file = "/Users/johnsolk/Documents/UCDavis/Whitehead/clust_results/Orthogroups_reformatted.csv"
orthogroups.to_csv(new_orthogroups_file,sep="\t",index=False,index_label=None)
```
